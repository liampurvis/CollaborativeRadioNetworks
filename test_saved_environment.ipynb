# File loading demo:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipywidgets import interact, widgets
import pickle
from os import listdir

def load_object(filename):
    with open(filename, 'rb') as  input:
        return pickle.load(input)


```

```python
data = pd.DataFrame()

for experiment in listdir("saved_environments/"):
    for simulation in listdir("saved_environments/" + experiment + "/"):
        for player in load_object("saved_environments/" + experiment + "/" + simulation):
            data = pd.concat([data, pd.DataFrame([[experiment, simulation, 
                                                   player[0], player[1],
                                                   np.sum(player[2]), np.sum(player[2][1200:1500])]], 
                                                 columns=["experiment_id", "simulation_id", 
                                                         "player_id", "player_type",
                                                         "player_success", "success_last_300_steps"])], 
                             axis = 0)
data
```
